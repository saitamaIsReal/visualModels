# FSII Analysis with ViT and 7×7 Patch Grid

This notebook demonstrates how to:
1. Divide an input image into a 7×7 patch grid and apply masking.  
2. Compute the model logits for various masked image variants.  
3. Approximate 2nd-order Faithful Shapley Interaction Indices (FSII) using SHAPIQ.  
4. Visualize the top-5 patch-pair interactions on the image.


````python
# 1 Imports & Monkey-Patch to disable the Border-Trick
import shapiq.approximator.sampling as _sampling

def _no_border(self, sampling_budget: int) -> int:
    # never apply the border-trick, just return the budget
    return sampling_budget

_sampling.CoalitionSampler.execute_border_trick = _no_border

import sys, os
import torch
import numpy as np
from transformers import ViTImageProcessor, ViTForImageClassification
from PIL import Image, ImageDraw
import requests
from io import BytesIO
from shapiq.approximator.regression import RegressionFSII

# 2) Global parameters
model_name   = "google/vit-base-patch32-384"
image_url    = "https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/cats.png"
image_size   = 384
grid_rows    = grid_cols = 7
n_patches    = grid_rows * grid_cols
device       = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print("Using device:", device)
if device.type == "cuda":
    print(" GPU:", torch.cuda.get_device_name(device))
````

## Load Image & Display 7×7 Patch Grid


````python
# Load and resize image
resp  = requests.get(image_url)
image = Image.open(BytesIO(resp.content)).convert("RGB").resize((image_size, image_size))

# Draw the 7×7 grid
cell_h, cell_w = image_size // grid_rows, image_size // grid_cols
grid_img = image.copy()
draw = ImageDraw.Draw(grid_img)
for r in range(grid_rows):
    for c in range(grid_cols):
        x1, y1 = c*cell_w, r*cell_h
        draw.rectangle([x1, y1, x1+cell_w, y1+cell_h], outline="gray", width=1)
        idx = r*grid_cols + c
        draw.text((x1+2, y1+2), str(idx), fill="gray")

grid_img
````

## Masking Function & Example Masks


````python
def mask_image_grid(img: Image.Image, coalition: np.ndarray) -> Image.Image:
    """Apply a gray mask to the patches flagged False in `coalition`."""
    arr = np.array(img.copy())
    coal = np.asarray(coalition, bool)
    for i, keep in enumerate(coal):
        if not keep:
            r, c = divmod(i, grid_cols)
            y1, y2 = r*cell_h, (r+1)*cell_h
            x1, x2 = c*cell_w, (c+1)*cell_w
            arr[y1:y2, x1:x2] = 128
    return Image.fromarray(arr)

# 1) Run a tiny sampling pass, return_interactions=True um die Koalitionen zu behalten
small_approx = RegressionFSII(
    n=n_patches,
    max_order=2,
    pairing_trick=False,
    sampling_weights=w,
    random_state=42
)
_ = small_approx.approximate(
    budget=50,             # kleines Demo-Budget
    game=value_fn,
    return_interactions=True
)

# 2) Greife auf die tatsächlich gezogenen Koalitionen zu
coalitions = small_approx._sampler.coalitions_matrix   # shape (n_samples, n_patches)

# 3) Zeige die ersten 6 Masken
from IPython.display import display
print("First 6 sampled coalitions (True=keep, False=mask):")
for idx in range(6):
    mask = coalitions[idx]
    print(f"‒ Sample #{idx:2d}, keeps {mask.sum()} patches")
    display(mask_image_grid(image, mask).resize((192,192)))
````

## Logit Checks (Full vs. Empty vs. Examples)


````python
# Load processor & model
processor = ViTImageProcessor.from_pretrained(model_name)
model     = ViTForImageClassification.from_pretrained(model_name).to(device).eval()

# Determine target class on the full image
inputs = processor(images=image, return_tensors="pt").to(device)
with torch.no_grad():
    logits = model(**inputs).logits
target_class = int(logits.argmax(-1))

def value_fn(coalitions: list[np.ndarray]) -> np.ndarray:
    out = []
    for coal in coalitions:
        masked = mask_image_grid(image, coal)
        batch = processor(images=masked, return_tensors="pt").to(device)
        with torch.no_grad():
            out.append(model(**batch).logits[0, target_class].item())
    return np.array(out)

# Full vs. Empty
print("Logit (full) :", value_fn([np.ones(n_patches)])[0])
print("Logit (empty):", value_fn([np.zeros(n_patches)])[0])

# Example masks
for idx, coal in enumerate(examples[2:], start=2):
    print(f"Mask {idx} logit:", value_fn([coal])[0])
````

````python

# Configure and run RegressionFSII
w = np.zeros(n_patches+1)
w[1], w[2] = 0.2, 0.8
w /= w.sum()

approximator = RegressionFSII(
    n=n_patches,
    max_order=2,
    pairing_trick=False,
    sampling_weights=w,
    random_state=42
)
res = approximator.approximate(
    budget=300,            # small demo budget
    game=value_fn,
    return_interactions=False
)

# Extract and print top-5 2nd-order interactions
fsii_map = res.dict_values
second = {p:v for p,v in fsii_map.items() if len(p)==2}
top5 = sorted(second.items(), key=lambda kv: abs(kv[1]), reverse=True)[:5]
for rank,(pair,val) in enumerate(top5,1):
    print(f"{rank}. {pair} → FSII = {val:.4f}")
````

**Notes**

- **Monkey-Patch** disables SHAPIQ’s internal “border-trick” to avoid dimension/exceeded errors.  
- **sampling_weights** controls the fraction of budget for single vs. pair masks (here 20 %/80 %).  
- **return_interactions=False** returns only the compact `dict_values` of final FSII scores (no raw samples).
